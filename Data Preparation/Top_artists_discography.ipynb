```python
import json
import requests
import pandas as pd
import numpy as np
```

```python
def format_clean(df,column):
    df['format_category']=np.where(df[column].str.contains("CD"),"CD",
                               np.where(df[column].str.contains("LP"),"LP",
                                        np.where(df[column].str.contains("DVD"),"DVD","Other")))
    return df
```

## Importing Top Artists discography from the discogs API

```python
#549 is Oren Ambarchi artist id
r = requests.get('https://api.discogs.com/artists/549/releases?page=1&per_page=500&sort=year') 
resp_json = r.json()
```

```python
data_ambarchi=json.loads(r.text)
```

```python
# Checking the size of the data
print(len(data_ambarchi['releases']))
```

```python
data_ambarchi=pd.json_normalize(data_ambarchi['releases'])
```

```python
format_clean(data_ambarchi,'format')
```
