```python
import pandas as pd
import numpy as np
```

```python
inventory=pd.read_csv('inventory.csv')
```

```python
def format_clean(df,column):
    df['music_format']=np.where(df[column].str.contains("CD"),"CD",
                               np.where(df[column].str.contains("LP"),"LP",
                                        np.where(df[column].str.contains("DVD"),"DVD","Other")))
    return df
```

```python
format_clean(inventory,'format')
```

```python
def date_conv(df,column):
    df['Date_listed']=pd.to_datetime(df[column])
    df['Date_listed']=df['Date_listed'].dt.date
    return df
```

```python
date_conv(inventory,'listed')
```

```python
inventory.to_csv('inventory_final.csv')
```
